```python
from pulp import *
# Personas y objetos a transportar
Fam = ['Mamá', 'Hija', 'Papá', 'Hijo']
Obj = ["Carpa","Cobija","Bolsa de dormir", "Linterna", "Balón","Botella de agua", "Lata de atún","Botiquín", "Lata de salchichas", "Repelente", "Protector solar", "Brújula"]
# preferencia de cada persona con cada objeto
pref = {('Papá','Botella de agua'):8,('Mamá','Botella de agua'):2,('Hijo','Botella de agua'):11,('Hija','Botella de agua'):8,
     ('Papá','Carpa'):2,('Mamá','Carpa'):9,('Hijo','Carpa'):5,('Hija','Carpa'):4,
     ('Papá','Lata de atún'):6,('Mamá','Lata de atún'):12,('Hijo','Lata de atún'):7,('Hija','Lata de atún'):3,
     ('Papá','Botiquín'):12,('Mamá','Botiquín'):4,('Hijo','Botiquín'):6,('Hija','Botiquín'):12,
     ('Papá','Protector solar'):11,('Mamá','Protector solar'):1,('Hijo','Protector solar'):12,('Hija','Protector solar'):7,
     ('Papá','Lata de salchichas'):7,('Mamá','Lata de salchichas'):10,('Hijo','Lata de salchichas'):10,('Hija','Lata de salchichas'):6,
     ('Papá','Bolsa de dormir'):3,('Mamá','Bolsa de dormir'):7,('Hijo','Bolsa de dormir'):3,('Hija','Bolsa de dormir'):10,
     ('Papá','Linterna'):10,('Mamá','Linterna'):8,('Hijo','Linterna'):2,('Hija','Linterna'):2,
     ('Papá','Repelente'):9,('Mamá','Repelente'):3,('Hijo','Repelente'):9,('Hija','Repelente'):9,
     ('Papá','Brújula'):5,('Mamá','Brújula'):5,('Hijo','Brújula'):1,('Hija','Brújula'):1,
     ('Papá','Cobija'):4,('Mamá','Cobija'):11,('Hijo','Cobija'):4,('Hija','Cobija'):11,
     ('Papá','Balón'):1,('Mamá','Balón'):6,('Hijo','Balón'):8,('Hija','Balón'):5
     }

# Capacidad de peso de cada persona
cap = {('Mamá'):12,
     ('Hija'):6,
     ('Papá'):12,
     ('Hijo'):6
     }

# Peso de cada objeto
peso = {('Botella de agua'):2,
     ('Carpa'):8,
     ('Lata de atún'):2,
     ('Botiquín'):2,
     ('Protector solar'):1,
     ('Lata de salchichas'):2,
     ('Bolsa de dormir'):4,
     ('Linterna'):3,
     ('Repelente'):2,
     ('Brújula'):1,
     ('Cobija'):6,
     ('Balón'):3
     } 

# Se define las variables X como binarias de 1 y 0 para ver si se tiene el objeto o no
binario = LpVariable.dicts("X", [(pers,cosa) for pers in Fam for cosa in Obj], 0, 1,LpBinary)

# Crear la funcion a maximizar, es decir la funcion de preferencias
funcion = LpProblem("Excursion",LpMaximize)
#print(funcion)

#definir exactamente como es la funcion a maximizar
funcion += lpSum(binario[pers,cosa]*pref[pers,cosa] for pers in Fam for cosa in Obj)

#Hay que llevar todos los objetos y no se pueden repetir
for cosa in Obj:
  funcion += lpSum(binario[pers,cosa] for pers in Fam) == 1

#No pueden llevar más peso del dado en la matriz de capacidad
for pers in Fam:
  funcion += lpSum(binario[pers,cosa]*peso[cosa] for cosa in Obj) <= cap[pers]

# Resolver el problema
funcion.solve()
print(funcion.solve)

# Función objetivo
print(f'Los puntos de preferencia obtenidos son: {value(funcion.objective)}')
print('')

for cosa in Obj:
  for pers in Fam:
    if binario[pers,cosa].varValue == 1:
      print(f'{pers} debe tener {cosa}')


```
